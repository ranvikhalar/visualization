```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL
import requests
import scipy
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

font1 = {'family': 'sans-serif', 'color' : 'red','size' : 15}
font2 = {'family': 'monospace', 'color' : 'green','size' : 10}

# Adding labels for each line plot
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')

# Adding legends with multiple options like location, frame on, fontsize, number of columns, and shadow
ax.legend(loc ='lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')

# bbox_to_anchor controls the placement of the legend relative to the bounding box of the axes or figure.
# (1,1) places the legend at the top-right corner, outside the plot.
# other options - (0,0) → Bottom-left, (1,0) → Bottom-right, (0.5, 1.1) → Slightly above the plot, etc.
ax.legend(loc ='upper left',
          bbox_to_anchor = (1, 1))
```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc ='upper left',
          bbox_to_anchor = (1, 0.5))

# Adding text to a specific point (10, 95)
ax.text(10, 95, 'This value is important')
```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc ='upper left',
          bbox_to_anchor = (1, 1))

# Adding text format color, size, ha (Horizontal Alignment). ha = 'right' aligns the right of the text to the data point. default value is left.
ax.text(10, 95, 'This value is important!',
        ha = 'right',
        color = 'red',
        size = 20)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Setting the axis limits
ax.axis([0, 10, 0, 10])
# Adding text annotations with different coordinate systems
# Data Coordinates. Places Data: (1, 5) at the data point (1,5)
ax.text(1, 5, "Data: (1,5)", transform = ax.transData)

# Axes Coordinates. Places text in axes coordinates (based on the axes' position in the plot area). In axes coordinates, (0, 0) is the bottom-left corner of the plot, and (1, 1) is the top-right corner.
# So, Axes: (0.5, 0.1) will be placed at bottom center position
ax.text(0.5, 0.1, "Axes: (0.5, 0.1)", transform = ax.transAxes)

# Places text in figure coordinates (relative to the whole figure, including all subplots).
# This places the text "Figure: (0.2, 0.2)" at the coordinates (0.2, 0.2) relative to the figure coordinates.
ax.text(0.2, 0.2, "Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Create a figure with two subplots (ax1 and ax2) arranged horizontally
fig, (ax1,ax2) = plt.subplots(ncols = 2, # 2 columns of subplots
                              nrows = 1, # 1 row of subplots
                              figsize = (7,3), layout = 'constrained') # Automatically adjust the layout to avoid overlap

font1 = {'family': 'sans-serif', 'color' : 'blue','size' : 20}
font2 = {'family': 'monospace', 'color' : 'green','size' : 14}

ax1.scatter(x1, y1, marker='*', color = 'indigo')
ax1.set_title('Scatter Plot', fontdict=font1, loc='center')
ax1.set_ylabel('Y-axis', fontdict=font1)
ax1.set_xlabel('X-axis', fontdict=font1)
ax1.annotate("(0, 50)", xy =(0, 50), xytext =(10,40),
            arrowprops=dict(facecolor='red', arrowstyle='wedge', shrinkA=60, shrinkB=7)) #shrinkA=5: Shrinks the tail of the arrow by 5 points from its starting position, shrinkB=5: Shrinks the arrowhead by 5 points from its endpoint.

ax2.bar(x2,y2, color = 'orange')
ax2.set_title('Bar Plot', fontdict=font2, loc='center')
ax2.set_ylabel('Y-axis', fontdict=font2)
ax2.set_xlabel('X-axis', fontdict=font2)
```

```python
# Create a figure with a custom subplot layout using the subplot_mosaic method
fig, someaxes =plt.subplot_mosaic([['ax1', 'ax3'], # First row: ax1 in the left column, ax3 in the right column
                                  ['ax2', 'ax3']], # Second row: ax2 in the left column, ax3 in the right column
                                  figsize = (7,4))

# Plot a scatter plot on the ax1 subplot
someaxes['ax1'].scatter(x1,y1)

# Plot a bar chart on the ax2 subplot
someaxes['ax2'].bar(x2,y2)

# Plot a line plot on the ax3 subplot
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
fig, someaxes =plt.subplot_mosaic([['ax1', 'ax3'],
                                  ['ax2', 'ax3']],
                                  figsize = (7,4),
                                  layout = 'constrained') # Automatically adjust the layout to avoid overlap. Another option for layout is tight

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7,3))

# Create both the Bar chart and the line plot on the same ax subplot
ax.bar(x, y1, color ='indigo')
ax.plot(x, y2, color  = 'red')
```

```python
# Calculate the standard deviation of the y2 data (used for the error bars)
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')

# Add error bars to the plot
ax.errorbar(x,
            y2,
            yerr = y2_sd, # Error value: standard deviation for each y value
            fmt ='none', # No marker for the data points
            ecolor = 'indigo', # Error bars color
            elinewidth = 2, # Width of the error bars
            capsize = 6, # Size of the caps at the end of the error bars
            capthick = 4, # Thickness of the caps
            errorevery = 2)  # Show error bars every 2nd point
```
