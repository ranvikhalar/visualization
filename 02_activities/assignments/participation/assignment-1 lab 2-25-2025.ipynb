```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613) # Reproducibility. Generate same results for each run

# Creating array with numbers 1 to 50, excluding 50
x = np.arange(50)

# Creating array of 50 random numbers starting from 0 to 100, excluding 100
y = np.random.randint(0, 100, 50)

# defining figure and axes with fig size (5,3)
fig, ax = plt.subplots(figsize = (5,3))

# Scatter Plot
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Bar Plot
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Line Plot
ax.plot(x,y)
```

# Activity

```python
fig, ax = plt.subplots(figsize = (5,3))

# Histogram
ax.hist(y, bins=10)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

# Adding Title and axis labels
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# defining font dictionaries
font1 = {'family': 'sans-serif', 'color' : 'blue','size' : 20}
font2 = {'family': 'monospace', 'color' : 'green','size' : 14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

# location settings
ax.set_title('Total growth over time', fontdict=font1, loc='right')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Adding marker and the color to the scatter plot
ax.scatter(x,y, marker='*', color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Adding linestyle, the linewidth, and the color code to the line plot
ax.plot(x,y, marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

# Adding markersize, markeredgecolor, and markerfacecolor to the line plot
ax.plot(x,y, 
        marker='o', 
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markersize = 10, 
        markeredgecolor = '#fa9359', 
        markerfacecolor = '#000000')

# Adding Grid lines. Default for both x & y. However, can be specified axis = 'x' or 'y'
ax.grid(color = 'red', linestyle = 'dotted', linewidth = 1.5)
```

```python

# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# create data
df = pd.DataFrame({
      'x': np.random.rand(40),
      'y': np.random.rand(40),
      'z': np.random.rand(40),
      'label': [f'Point {i}' for i in range(40)]
   })
df.head()

plt.scatter(df['x'], df['y'], s=df['z']*1000, alpha=0.5, color = 'red')
plt.title('Bubble Chart Example', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.show()
```

```python
import plotly.express as px

# Create data
np.random.seed(42)
df = pd.DataFrame({
    'x': np.random.rand(40),  
    'y': np.random.rand(40),
    'z': np.random.rand(40) * 100,  # Scale 'z' to avoid tiny bubbles
    'label': [f'Point {i}' for i in range(40)]  # Custom labels for hover info
})

# Create interactive scatter plot
fig = px.scatter(df, 
                 x='x', 
                 y='y', 
                 size='z',  # Bubble size
                 color='z',  # Color based on 'z' values
                 hover_name='label',  # Show label on hover
                 title='Interactive Bubble Chart')

# Show figure
fig.show()
```
