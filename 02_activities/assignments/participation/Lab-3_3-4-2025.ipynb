```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy

# For opening and processing image files
from PIL import Image

# For making HTTP requests to fetch the image
import requests

# For handling byte-stream data
from io import BytesIO
```

```python
x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([170,100,90,120,50])

# Sending an HTTP GET request to download the image from the provided URL
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')

# Creating a BytesIO object to store the image data in memory
image_file = BytesIO(response.content)

# Opening the image using PIL for further processing or display
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color ='red')

# Adding a new set of axes to the existing figure for displaying an image
# [left, bottom, width, height] → Defines the position and size of the new axes
# 0.78 → Left position (relative to figure width)
# 0.17 → Bottom position (relative to figure height)
# 0.15 → Width of the new axes
# 0.35 → Height of the new axes
ax_image = fig.add_axes([0.78, 0.17, 0.15, 0.35])

# Displaying the image inside the newly created axes
ax_image.imshow(image)

# Removing axis labels and ticks for a clean display
ax_image.axis('off')
```
