```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a 1x3 subplot layout without titles for each plot
fig = make_subplots(
    rows=1, 
    cols=3, 
)

# Scatter + Line Plot (Production)
fig.add_trace(
    go.Scatter(
        x=df['Year'],  # x-axis data
        y=df['Production (kilolitres)'],  # y-axis data
        mode='lines+markers',  # Displaying both lines and markers
        marker=dict(size=10, color='hotpink', opacity=1, line=dict(width=2, color='purple')),
        line=dict(color='blue', width=2)
    ),
    row=1, col=1  # Place this plot in the first subplot
)

# Scatter + Line Plot (Farm Value Per Litre)
fig.add_trace(
    go.Scatter(
        x=df['Year'],  # x-axis data
        y=df['Farm Value Per Litre ($/litre)'],  # y-axis data
        mode='lines+markers',  # Displaying both lines and markers
        marker=dict(size=10, color='green', opacity=1, line=dict(width=2, color='purple')),
        line=dict(color='blue', width=2)
    ),
    row=1, col=2  # Place this plot in the second subplot
)

# Scatter + Line Plot (Total Farm Value)
fig.add_trace(
    go.Scatter(
        x=df['Year'],  # x-axis data
        y=df["Total Farm Value ($'000)"],  # y-axis data
        mode='lines+markers',  # Displaying both lines and markers
        marker=dict(size=10, color='orange', opacity=1, line=dict(width=2, color='purple')),
        line=dict(color='blue', width=2)
    ),
    row=1, col=3  # Place this plot in the third subplot
)

# Update layout with axis titles and styling
fig.update_layout(
    title="Yearly Overview of Ontario Maple Syrup Production, Value Per Litre, and Total Farm Value",
    title_font=dict(size=25, color='darkblue', family='Arial',  # Title font color and bold styling
                    weight='bold'),
                    title_x=0.5,  # Centering the main title
    showlegend=False,  # Remove legend if not needed
    height=600,  # Set the height of the figure
    width=1200,  # Set the width of the figure
    # X and Y axis styles for all plots
    xaxis=dict(
        title="Year",
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # X-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    ),
    yaxis=dict(
        title="Production (kilolitres)",  # Y-axis title for the first plot
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # Y-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    ),
    xaxis2=dict(
        title="Year",
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # X-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    ),
    yaxis2=dict(
        title="Farm Value Per Litre ($/litre)",  # Y-axis title for the second plot
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # Y-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    ),
    xaxis3=dict(
        title="Year",
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # X-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    ),
    yaxis3=dict(
        title="Total Farm Value ($'000)",  # Y-axis title for the third plot
        title_font=dict(size=18, color='black', family='Arial', weight='bold'),  # Y-axis title color and bold styling
        showgrid=True,  # Show gridlines
        gridcolor='lightgray'  # Gridline color
    )
)

# Show the plot
fig.show()
```
